```python
from scipy import interpolate 
```

```python
t1 = np.linspace(0, 1, len(rn_lat))
n = 200
t_new = np.linspace(0, 1, n)
f = interpolate.interp1d(t1, rn_lat,kind='cubic')
rn_lat_new = f(t_new)
```

```python
t2 = np.linspace(0, 1, len(rn_lon))
n = 200
t_new = np.linspace(0, 1, n)
f = interpolate.interp1d(t1, rn_lon,kind='cubic')
rn_lon_new = f(t_new)
```

```python
#rn_lat = percorso2['Latitude'].values.tolist()
#rn_lon = percorso2['Longitude'].values.tolist()
points2 = np.zeros((len(rn_lon_new),2))
points2[0,:] = ([rn_lat_new[0],rn_lon_new[0]])
for i in range(len(rn_lat_new)):
        points2[i,:] = ([rn_lat_new[i],rn_lon_new[i]])
```

```python
rdc=pd.read_csv('/Users/davidelibera/Desktop/congo.csv')
```

```python
rdc=rdc.set_index('Year')
```

```python
rdc['Population'] = rdc['Population'].div(1000000)
rdc['Growth'] = rdc['Growth'].div(1000000)
```

```python
rdc
```

```python
fig,ax1 = plt.subplots(figsize=(14,6)) 
ax1.plot(rdc['Population'], marker='1', linestyle='-', linewidth = 1, label='Population', color='#f5d112')
ax1.plot(rdc['Growth'],marker='2', linestyle='-', linewidth = 0.4, label='Net Growth', color='#c20016')
ax2 = ax1.twinx() 
ax2.plot(rdc['GrowthRate'],marker='3', linestyle='-', linewidth = 0.4, label='Growth Rate', color='#1066ff')
plt.xticks(ticks=range(1950,2040,5))
ax1.grid(alpha=0.4)
ax2.grid(alpha=0.4)
ax1.set_ylabel('Population trend',size=14)
ax2.set_ylabel('Population growth rate trend',size=14)
plt.title('Population in Kinshasa',size=18)
plt.axvline(x=2022,color='black',linewidth=0.8)
ax1.set_facecolor('#f4f1ee')
ax1.text(2023,25,'2',fontsize=18)
ax1.text(2023,23,'0',fontsize=18)
ax1.text(2023,21,'2',fontsize=18)
ax1.text(2023,19,'2',fontsize=18)
ax1.legend()
ax2.legend(bbox_to_anchor=(0.001,0.83),loc=(0.1,0.1))
plt.show()
```
